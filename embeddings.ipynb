```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    "Artificial intelligence is transforming the way we interact with technology.",
    "Machine learning models improve their performance through experience.",
    "Deep learning uses neural networks to solve complex problems.",
    "Python is a popular programming language for data science and AI.",
    "Training a model requires clean and well-prepared data.",
    "Evaluation metrics help us understand how well a model performs.",
    "Natural Language Processing enables machines to understand human language.",
    "Deploying AI models in real-world applications requires careful planning and monitoring."
]
```

```python
sentences
```

```python
##Vocabulary size
voc_size=10000
```

```python
one_hot_rep=[one_hot(words , voc_size) for words in sentences]
one_hot_rep
```

```python
max=0
for sent in one_hot_rep:
    length=len(sent)
    if length>max:
        max=length
print(max)
```

```python
##Word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length = 12 
embedded_doc=pad_sequences(one_hot_rep  , padding='pre' , maxlen=sent_length)
embedded_doc
```

```python
## Feature representation
dim=10
```

```python
print(voc_size)
print(dim)
print(sent_length)
```

```python
model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_length=sent_length))
model.compile(optimizer='adam', loss='mse')
model.build(input_shape=(None, sent_length))  # <-- Optional but forces the model to build if not built yet
model.summary()
```

```python
model.predict(embedded_doc)
```
